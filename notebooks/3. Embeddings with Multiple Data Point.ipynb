```python
import numpy as np
from sentence_transformers import SentenceTransformer

# 1. Load a pretrained Sentence Transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
sentences = [
    "The weather is lovely today.",
    "It's so sunny outside!",
    "He drove to the stadium.",
]

dataset = []

for i, x in enumerate(sentences):
    dataset.append({
        'id': i,
        'text': x
    })
```

```python
print(dataset)
```

```python
embeddings = []

for x in dataset:
    embedding = model.encode(x.get('text'))
    embeddings.append((
        x.get('id'),
        embedding
    ))
```

```python
print(embeddings)
```

```python
query = "The weather is lovely today"
query_embedding = model.encode(query)

results = []
for text in embeddings:
    text_id = text[0]
    text_embedding = text[1]
    rank = model.similarity(text_embedding, query_embedding)
    results.append((
        text_id,
        rank
    ))
results
```

```python
results.sort(key=lambda x: x[1], reverse=True)
```

```python
for result in results:
    dataset_id = result[0]
    rank = result[1]
    text = next(data for data in dataset if data['id'] == dataset_id)
    print(dataset_id, rank, text)
```
